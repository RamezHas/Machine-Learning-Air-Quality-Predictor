```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../outputFiles/cleaned_air_quality.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
df.head()
```

```python
print("\nSummary Statistics:")
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["pm25"], bins=30, kde=True ,color='purple')
plt.title("Distribution of PM2.5 Levels")
plt.xlabel("Timestamp")
plt.ylabel("PM2.5")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["timestamp"], df["pm25"],marker='o',linestyle='-',color='teal')
plt.title("PM2.5 over Time")
plt.xlabel("Date")
plt.ylabel("PM2.5")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["pm25"],color='orange')
plt.title("PM2.5 Boxplot(Outlier Detection")
plt.xlabel("PM2.5")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.scatterplot(data=df, y="pm25", x="temperature", color="blue", ax=axes[0])
axes[0].set_title("PM2.5 vs Temperature")
axes[0].set_xlabel("Normalized Temperature")
axes[0].set_ylabel("PM2.5")
sns.scatterplot(data=df,x="humidity",y="pm25",color="green",ax=axes[1])
axes[1].set_title("PM2.5 vs. Humidity")
axes[1].set_xlabel("Normalized Humidity")
axes[1].set_ylabel("PM2.5")
plt.tight_layout()
plt.show()
```
